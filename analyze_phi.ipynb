```python
import numpy as np
import pandas as pd
```

```python
test = pd.read_csv('test_phi_coefficients+node2.csv')
train = pd.read_csv('train_phi_coefficients+node2.csv')

y_c_name = np.array(train['Y_c_name'])
```

```python
for i in range(40):
    start_idx = i*40
    print('SkewRank: {} {}'.format(i, y_c_name[start_idx]))
    def get_top_three(data):
        y_o_name = np.array(data['Y_o'])
        phi = np.array(data['Phi'])
        output = ''
        for j in range(start_idx, start_idx+6):
            if y_o_name[j]==y_c_name[j]: continue
            output += '{}({:.2f}), '.format(y_o_name[j], phi[j])
        return output
    output = '_ phi train: ' + get_top_three(train) + '\n'
    output += '_ phi test: ' + get_top_three(test) + '\n'
    
    
    print(output)
```
